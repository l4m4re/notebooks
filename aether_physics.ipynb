### Definition of the medium

We assume all space to be filled with a superfluid like medium, which behavior can be defined by three constants: the speed of licht, the viscosity of the medium and Planck's constant. 

Hereby, we define the value of viscosity $\eta$ by the inverse of vacuum permeability  $\mu_0$, which used to be defined as:  $4\pi \times 10^{-7}$ H/m.

(Ref: [Wikipedia](https://en.wikipedia.org/wiki/Vacuum_permeability) )


----

```python
from math import *

c     = 299792458.0     # speed of light     [m/s]
eta   = 1/(4*pi*1e-7)   # viscosity          [kg/m-s],   [Pa-s] 
h     = 6.62607015e-34  # Planck's constant  [kg-m^2/s], [J-s]  
```

From here, we can define further constants, starting with mass density $\rho$, which we initially define to have a value equal to permittivity ${\epsilon_0}$, but a unit of measurement in $[kg/m^3]$ by using the definition of permittivity ${\epsilon_0}$:

$$
\displaystyle {\epsilon_0 = \frac{1}{c^2 \mu_0}}.
$$
<p style="text-align: right;">(eq. 3)</p>

And substituting $\rho = \epsilon_0$ and $\eta = 1/\mu_0$:

$$
\displaystyle {\rho = \frac{\eta}{c^2}}. 
$$
<p style="text-align: right;">(eq. 4)</p>

This comes within 10 significant digits of the SI value for permittivity ${\epsilon_0}$, also known as the electric constant. The reason for this is that the susceptibility ${\mu_0}$ has been redefined in 2019, while we prefer the classical exact definition.

From here, we can work out the unit of measurement for charge in Coulomb by equating the units of measurement of $\rho$ in $[\frac{kg}{m^3}]$ and $\epsilon_0$  in $[\frac{C^2}{N-m^2}]$ or $[\frac{C^2 \cdot s^2}{kg \cdot m^3}]$:

$$
\frac{C^2 \cdot s^2}{kg \cdot m^3} = \frac{kg}{m^3} \Rightarrow C^2 \cdot s^2 = kg^2 \Rightarrow C = \frac{kg}{s}
$$
<p style="text-align: right;">(eq. 5)</p>

Thus, charge in Coulombs can be viewed as having the unit $[\frac{kg}{s}]$, providing a new physical dimension to the concept of charge.


Then, we can define the diffusivity, kinematic viscosity or quantum circulation constant $k$, as follows:

$$
\displaystyle k = \frac{\eta}{\rho}. 
$$
<p style="text-align: right;">(eq. 6)</p>

This way, we obtain a unit of measurement in $[m^2/s]$ and a value equal to light speed c squared for $k$. 

However, because $c$ currently has an exact definition, we change the order of these definitions a bit and define $k$ as having a value equal to $c^2$ and a unit of measurement in $[m^2/s]$ and use that to define mass density $\rho$ as follows:

$$
\displaystyle \rho  = \frac{\eta}{k}. 
$$

<p style="text-align: right;">(eq. 7)</p>

Fundamentally, neither light speed $c$, quantum circulation $k$ nor mass density $\rho$ can be considered absolute Universal constants, as argued by a/o Hatch. But it appears that viscosity $\eta$ may be an actual Universal constant, given it's classical exact definition that includes $\pi$, suggesting it actually represents some kind of geometric proportion.

```python
k   = c**2  # quantum circulation constant
            # 8.987551787368176e+16                 [m^2/s]
rho = eta/k # mass density   8.85418781762039e-12   [kg/m^3]    

print("c   = ", c)
print("eta = ", eta)
print("rho = ", rho)
print("k   = ", k)
```

### Vector Laplacian

The vector Laplacian plays a crucial role in our model and is defined by:

$$
\displaystyle \Delta \vec F = \text{grad div} \vec F - \text{curl curl} \vec F. 
$$

<p style="text-align: right;">(eq. 10)</p>

(Ref. [Wikipedia](https://en.wikipedia.org/wiki/Laplace_operator#Vector_Laplacian) )

The vector Laplace operator describes the second spatial derivative in three dimensions, which would be the three dimensional generalization of the one dimensional $d^2/dx^2$, and has a unit of measurement in per meter squared $[/m^2]$. 

What the Laplace operator does is establish a Helmholtz decomposition of the vector field $\vec F$, whereby the term grad div $\vec F$ results in a **linear** component, the curl free or irrotational field usually denoted $\vec E$, and the term curl curl $\vec F$ results in an **angular** component, the divergence free or incompressible field usually denoted $\vec B$. And these two components may be evaluated independently of one another. 

(Ref. [Wikipedia](https://en.wikipedia.org/wiki/Helmholtz_decomposition) )

It is because of this resulting Helmholtz decomposition that the vector Laplacian actually defines a fundamental symmetry between these linear and angular components and it is that fundamental symmetry that is broken in theories that do not derive their fields directly using the vector Laplacian, which includes Maxwell's equations as well as Navier-Stokes.

The div operator in the vector Laplacian is defined as a surface integral over a
spherical volume $(4/3) \pi R^3$, divided by radius R, whereby the limit of R is
taken to zero, while the grad operator returns the change in a scalar field $f$
over a displacement dR, divided by the distance dR. 

Similarly, the curl operator also has a unit in [$/m$] and is defined as a
line integral over a closed curve, divided by the area enclosed by that curve,
whereby the limit of the area is taken to zero.

It is important to note that the div operator involves an integral over the
surface of a volume, and therefore works on a **volume** rather than a surface, while
the curl operator involves a line integral over a closed curve, and therefore
works on a **surface** rather than a volume. 

The difference between these volumetric and planar operations is crucial in the
understanding of the geometric aspects of the vector Laplacian, which becomes
clear when implementing a fdtd simulator using a staggered Yee grid. In our work
on a simulator, we found that it is necessary to implement two distinct
versions of the curl operator and that there is a fundamental difference between
volumetric "point" fields and "surface" fields.

(Ref. [Operators.py](https://github.com/l4m4re/fdtd/blob/master/fdtd/operators.py ))

In other words, the discrete finite difference implementation of the vector
Laplacian learns that the various fields in the model are located at specific
geometric points in space relative to one another, a detail that is lost in
infinitesimal calculus.


### Field Definitions

In the following, we will define the fields in our model, whereby we use the symbol $\Gamma$ for the circulation in line with fluid dynamics. For the aether this circulation $\Gamma$ is equal to the quantum circulation constant $k$.

We can define both scalar and vector potentials by working out the vector Laplacian for the velocity field, multiplied by the circulation constant, to compute the acceleration field:

$$\vec{a} = - \Gamma \Delta \vec{v}$$
<p style="text-align: right;">(eq. 11)</p>

Which, considering the definition of the vector Laplacian $\Delta$, works out to:

$$
\vec{a} = \Gamma \left( -\nabla \nabla \cdot \vec{v} 
        + \nabla \times \nabla \times \vec{v} \right),
$$
<p style="text-align: right;">(eq. 12)</p>

We can now define fields similar to the familiar scalar potential $\Phi$, vector
potential $\vec A$ and the familiar $\vec E$ and $\vec B$ fields. 

However, in order to maintain consistency, we deviate from the traditional
definition for the vector potential, both in electrodynamics as well as fluid
dynamics, where the vector potential is defined by:

$$
\Delta\vec A= - \nabla \times ( \nabla \times \vec A ) = - \vec \omega .
$$
<p style="text-align: right;">(eq. 13)</p>

(Ref. Richardson, S., & Cornish, A. (1977). Solution of three-dimensional
incompressible flow problems. Journal of Fluid Mechanics, 82(2), 309-319.
[doi:10.1017/S0022112077000688](https://pdfs.semanticscholar.org/9344/48b028a3a51a7567c2b441b5ca3e49ebb85c.pdf) (eq. 19)
)

And, in order to avoid confusion, we choose to denote the linear scalar
potential by $p$ for pressure, the linear force density field by $\vec f$ and
the angular torque density field by $\vec \tau$, instead.

We can now define the scalar potential by:

$$
p = \rho \Gamma \nabla \cdot \vec{v} =  \eta \nabla \cdot \vec{v},
$$
<p style="text-align: right;">(eq. 14)</p>

resulting in a unit of measurement of **pressure** in $[Pa]$ or $[N/m^2]$ representing **surface force density**.

We can then define the linear **force density** field $\vec{f}$, in $[N/m^3]$ by:

$$
\vec{f} = - \nabla p = - \eta \nabla  \nabla \cdot \vec{v} .
$$
<p style="text-align: right;">(eq. 15)</p>

And from here, we can define the linear acceleration field $\vec{a_l}$ by:

$$
\vec{a_l} = - \frac{1}{\rho} \nabla p 
          =   \frac{1}{\rho} \vec f 
          = - \Gamma \nabla  \nabla \cdot \vec{v} .
$$
<p style="text-align: right;">(eq. 16)</p>


And since in fluid dynamics, the vorticity ${\vec \omega}$ is defined as the curl of the velocity field $\vec{v}$:

$$ \vec{\omega} = \nabla \times \vec{v} , $$ 
<p style="text-align: right;">(eq. 17)</p>

we can define the angular counterpart of force density $\vec f$, torque density $\vec \tau$, by:

$$
\vec{\tau} = \eta \vec{\omega} 
           = \eta \nabla \times \vec{v} ,
