## Abstract

It is shown that from the assumption that all space is filled with a superfluid medium called aether, characterized by the quantum circulation constant $k$, an elemental particle can be described, characterized by a mass $m={h \over k}$, with $h$ Planck's constant, and a charge $q*q = e*e$, with e elemental charge.

It is shown that from these equations, both the toroidal radius $R$ as well as the hollow core radius $a$ of such an elemental vortex ring particle can be computed, as well as their ratio $\epsilon = {a \over R}$. 

From there, it is shown that by starting from the definition of the fine structure constant $\alpha$, the definition of elementary charge $e*e$ can be rewritten by multiplying by $m/m$, $R/R$ and $\pi/\pi$ such that two terms are obtained, both of which are exactly equal in magnitude to elemental charge $e$, whereby one of the terms represents an angular or magnetic component and the other represents a linear or electric component. And finally, it is shown that the fine structure constant $\alpha$ can be described as a function of $\epsilon$. 

Thus, it is shown that the heretofore unexplained nature of the fine structure constant $\alpha$ can be attributed to the coupling of the linear and angular components of the charge of such an elemental ring vortex, offering a deep understanding of the actual nature of charge to which heretofore a more or less arbitrary unit of measurement, the Coulomb, was assigned. ## Vector Laplace operator

Next, we consider the vector Laplace operator, defined by:

$$
\displaystyle Δ \vec F = \text{grad div} \vec F - \text{curl curl} \vec F. 
$$

<p style="text-align: right;">(eq.1)</p>

(Ref. [Wikipedia](https://en.wikipedia.org/wiki/Laplace_operator#Vector_Laplacian) )

The vector Laplace operator describes the second spatial derivative in three dimensions, which would be the three dimensional generalization of the one dimensional $d^2/dx^2$, and has a unit of measurement in per meter squared [$/m^2$]. What is very interesting is that together with the quantum circulation constant $k$, we can actually define the time derivative of any given vector field in the physical three dimensional space-time as follows:

$$
\displaystyle {d \vec F \over dt} = - k Δ \vec F. 
$$
<p style="text-align: right;">(eq.2)</p>

Obviously, the units of measurements match: $[/s] = [m^2/s] [/m^2]$.

This is very significant, because it enables us to define higher order Laplace and Poisson equations in three dimensions and it also gives us the opportunity to obtain a much deeper understanding of the Nature of space and time in our Universe. In other words: this is the fundamental equation that will one day be recognized as one of the biggest scientific breakthroughs of the 21st century, because there is simply no argument whatsoever to be made against such a simple and straightforward application of the vector Laplacian.

What the Laplace operator does is establish a Helmholtz decomposition of the vector field $\vec F$, whereby the term grad div $\vec F$ results in a **linear** component, the curl free or irrotational field usually denoted $\vec E$, and the term curl curl $\vec F$ results in an **angular** component, the divergence free or incompressible field usually denoted $\vec B$. And these two components may be evaluated independently of one another. 

(Ref. [Wikipedia](https://en.wikipedia.org/wiki/Helmholtz_decomposition) )

When we consider the ratio of mass to mass density $m/ρ$, we obtain a unit of measurement in $[kg] [m^3/kg] = [m^3]$,which describes a volume V. And since the vector Laplacian has a unit in [$/m^2$] we obtain a unit of measurement in [$m$] or displacement when we compute Δ $m/ρ$. 

The grad div operator in the vector Laplacian can be thought of as a surface integral over a spherical volume $(4/3) π R^3$, divided by radius R, whereby the limit of R is taken to zero. However, we can compute the same operation without taking the limit of R to zero by computing the radius to volume ratio R/V for such an object, such as a ring vortex with big toroidal radius R and small poloidal radius r, and then use multiplication, provided the fields involved are symmetrical with respect to the surface area of the object. 

Similarly, the curl curl operator also has a unit in [$/m^2$] we obtain a unit of measurement in [$m/s$] or velocity when we compute curl curl $e/ρ$, which represents a tangential velocity along some closed curve.

Since in our model charge has a unit of measurement in [$kg/s$], when multiplying Δ$m/ρ$ with charge q, we obtain a unit of measurement in [$kg-m/s$] or momentum. This way, we can define momentum $P$ for a ring vortex with mass m and charge q by:

$$
\displaystyle P = q ( \text{ grad div }\text (m/ρ) + \text{ curl curl }\text (m/ρ) ), 
$$
<p style="text-align: right;">(eq. 5)</p>

which contains both a linear and an angular component.

Next, we consult eq. 1.9 in "dynamics of thin vortex rings", describing Fraenkel’s second-order formula for energy $E$:

$$
\displaystyle E = {1\over2} ρ R Γ^2 \left( ln {8 \over ε} − {7 \over 4} + {3 \over 16} ε^2 ln {8 \over ε} \right).
$$
<p style="text-align: right;">(eq. 6)</p>

And compare to eq. 1.5:

$$
\displaystyle E = {1\over2} ρ R Γ^2 \left( ln {8 \over ε} − α  \right),
$$
<p style="text-align: right;">(eq. 6)</p>

with α some constant, not to be confused with the fine structure constant. From these, we can define this α:

$$
\displaystyle α = {7 \over 4} - {3 \over 16} ε^2 ln {8 \over ε}
$$

<p style="text-align: right;">(eq. 6)</p>

Next, we consult eq 1.7:

$$
\displaystyle v = {∂E \over ∂P} = {Γ \over 4πR} \left(ln {8R \over a} + 1 − α \right)
$$
<p style="text-align: right;">(eq. 6)</p>

Substitute ε=R/a:

$$
\displaystyle v = {∂E \over ∂P} = {Γ \over 4πR} \left(ln {8 \over ε} + 1 − α \right)
$$
<p style="text-align: right;">(eq. 6)</p>

Substitute $α = {7 \over 4} - {3 \over 16} ε^2 ln {8 \over ε}$:

$$
\displaystyle v = {∂E \over ∂P} = {Γ \over 4πR} \left(ln {8 \over ε} + 1 − ({7 \over 4} - {3 \over 16} ε^2 ln {8 \over ε}) \right)
$$
<p style="text-align: right;">(eq. 6)</p>

And simplify:

$$
\displaystyle v = {∂E \over ∂P} = {Γ \over 4πR} \left(ln {8 \over ε} - {3 \over 4} + {3 \over 16} ε^2 ln {8 \over ε} \right)
$$

$$
\displaystyle v = {∂E \over ∂P} = {Γ \over 4πR} \left(ln {8 \over ε} + {3 \over 16} ε^2 ln {8 \over ε} - {3 \over 4} \right)
$$

$$
\displaystyle v = {∂E \over ∂P} = {Γ \over 4πR} \left(ln {8 \over ε} (1 + {3 \over 16} ε^2) - {3 \over 4} \right)
$$
<p style="text-align: right;">(eq. 6)</p>

And equate

$$
\displaystyle v = {1 \over (2 π^2 r^2)} {q \over ρ} = {Γ \over 4πR} \left(ln {8 \over ε} (1 + {3 \over 16} ε^2) - {3 \over 4} \right)
$$
<p style="text-align: right;">(eq. 6)</p>

substitute r = a = εR and $k=Γ$:

$$
\displaystyle v = {1 \over (2 π^2 ε^2R^2)} {q \over ρ} = {k \over 4πR} \left(ln {8 \over ε} (1 + {3 \over 16} ε^2) - {3 \over 4} \right)
$$
<p style="text-align: right;">(eq. 6)</p>

Solve for q:

$$
\displaystyle  q = {k ρ  2 π^2 ε^2R^2 \over 4πR} \left(ln {8 \over ε} (1 + {3 \over 16} ε^2) - {3 \over 4} \right)
$$
<p style="text-align: right;">(eq. 6)</p>

Simplify:

$$
\displaystyle  q = {k ρ  π  R} * {{ε^2} \over 2} \left(ln {8 \over ε} (1 + {3 \over 16} ε^2) - {3 \over 4} \right)
$$
<p style="text-align: right;">(eq. 6)</p>

And subsitute $k=h/m$

$$
\displaystyle  q = ρ {h \over m} π R * {{ε^2} \over 2} \left(ln {8 \over ε} (1 + {3 \over 16} ε^2) - {3 \over 4} \right)
$$
<p style="text-align: right;">(eq. 6)</p>

And thus we can update our correction factor γ to:

$$
\displaystyle γ = {{ε^2} \over 2} \left(ln {8 \over ε} (1 + {3 \over 16} ε^2) - {3 \over 4} \right)
$$
<p style="text-align: right;">(eq. 10)</p>



```python
import numpy as np
from scipy.optimize import fsolve

# Define the constants
q     = 1.602176634e-19
h     = 6.62607015e-34
k     = 8.987551787368176e+16
eta   = 1/(4*np.pi*1e-7)   # viscosity          [kg/m-s],   [Pa-s] 
rho   = eta/k   # mass density   8.85418781762039e-12   [kg/m^3]

Rin   = 6.408706536e-26
epsIn = 0.1
#Rin   = 1.0597925343886396e-26
#epsIn = 2.594130071505027

def q_l(vars):
    R, eps = vars
    
    return k * rho * np.pi * eps**2 * R / 2  \
             * ( np.log(8/eps)               \
                 * (1 + 3/16 * eps**2) - 3/4 \
               )   
             
def q_a(vars):
    R, eps = vars
    
    return k * rho * np.pi * R               \
             * (1 + 3/4 * eps**2) 

# Define the equations to be solved
def equations(vars):
    R, eps = vars
    
    eq1 = q - q_l(vars)
    eq2 = q - q_a(vars)
      
    #print(q_l(vars), q_a(vars))    
        
    return [eq1, eq2]

# Solve the equations
R, eps = fsolve(equations, [Rin, epsIn])

# Print the results
print(f"R   = {R}")
print(f"eps = {eps}")
print(f"q_l = {q_l([R, eps])}")
print(f"q_a = {q_a([R, eps])}")
print(f"e   = {e}")
```

## Corrected definition of elemental charge

Next step is to correct our definition for linear elementary charge by including the term $(1 + 3/4 ε^2)^2$ to come to our new definition of elemental charge squared:

$$
\displaystyle e*e = ρ {h \over m} π R * { m c \over 4 π^3 √3 R (1 + 3/4 ε^2)^2}.
$$
<p style="text-align: right;">(eq. 29)</p>

or, with a sneak preview of what's to come:

$$
\displaystyle e*e = ρ {h \over m} π R * { 2 m c α \over π R}.
$$
<p style="text-align: right;">(eq. 30)</p>

This illustrates that α actually represents a coupling factor between the linear (electric) and angular (magnetic) fields describing the dynamics of the elemental superfluid vortex ring in the medium.

And then we evaluate both terms again:

----

```python
a_e = 1 + (3/4) * eps *eps

linear  = (m*c)/(4*pi**3*sqrt(3)*R*a_e*a_e)
angular = rho*(h/m)*pi*R
q       = sqrt(linear*angular)

print("angular or magnetic charge: ", angular)
print("linear or dielectric charge:", linear)
print("elemental charge:           ", q)
```

## New definition of fine structure constant α

Next step is to return to the equation we started with:


$$
\displaystyle e^2 = e*e = {2 α ρ h c}.
$$
<p style="text-align: right;">(eq. 31)</p>

And equate this to our corrected equation:

$$
\displaystyle e*e = {2 α ρ h c} = ρ {h \over m} π R * { m c \over 4 π^3 √3 R (1 + 3/4 ε^2)^2}.
$$
<p style="text-align: right;">(eq. 32)</p>

And solve for α to obtain a new definition for α:

$$
\displaystyle α = { 1 \over 8 π^2 √3 (1 + 3/4 ε^2)^2}.
$$
<p style="text-align: right;">(eq. 33)</p>

We can compute this and compare to the the 2018 CODATA recommended value of α:

$$
\displaystyle α = 0.007297352569311
$$
<p style="text-align: right;">(eq. 34)</p>

(Ref. [Wikipedia](https://en.wikipedia.org/wiki/Fine-structure_constant) )

----

```python
alpha        = ((rho*h*R*R)/(2*c*m*m))*(1+(3/4)*eps*eps)
alpha_codata = 0.007297352569311

print("alpha:    ",alpha)
print("codata:   ",alpha_codata)
print("relative: ",alpha_codata/alpha)
```
