### Vector Laplacian

The vector Laplacian plays a crucial role in our model and is defined by:

$$
\displaystyle \Delta \vec F = \text{grad div} \vec F - \text{curl curl} \vec F. 
$$

<p style="text-align: right;">(eq. 8)</p>

(Ref. [Wikipedia](https://en.wikipedia.org/wiki/Laplace_operator#Vector_Laplacian) )

The vector Laplace operator describes the second spatial derivative in three dimensions, which would be the three dimensional generalization of the one dimensional $d^2/dx^2$, and has a unit of measurement in per meter squared [$/m^2$]. 

What the Laplace operator does is establish a Helmholtz decomposition of the vector field $\vec F$, whereby the term grad div $\vec F$ results in a **linear** component, the curl free or irrotational field usually denoted $\vec E$, and the term curl curl $\vec F$ results in an **angular** component, the divergence free or incompressible field usually denoted $\vec B$. And these two components may be evaluated independently of one another. 

(Ref. [Wikipedia](https://en.wikipedia.org/wiki/Helmholtz_decomposition) )

It is because of this resulting Helmholtz decomposition that the vector Laplacian actually defines a fundamental symmetry between these linear and angular components and it is that fundamental symmetry that is broken in theories that do not derive their fields directly using the vector Laplacian, which includes Maxwell's equations as well as Navier-Stokes.

The div operator in the vector Laplacian is defined as a surface integral over a
spherical volume $(4/3) \pi R^3$, divided by radius R, whereby the limit of R is
taken to zero, while the grad operator returns the change in a scalar field $f$
over a displacement dR, divided by the distance dR. 

Similarly, the curl curl operator also has a unit in [$/m^2$] and is defined as a
line integral over a closed curve, divided by the area enclosed by that curve,
whereby the limit of the area is taken to zero.

It is important to note that the div operator involves an integral over the
surface of a volume, and therefore works on a **volume** rather than a surface, while
the curl operator involves a line integral over a closed curve, and therefore
works on a **surface** rather than a volume. 

The difference between these volumetric and planar operations is crucial in the
understanding of the geometric aspects of the vector Laplacian, which becomes
clear when implementing a fdtd simulator using a staggered Yee grid. In our work
on a simulator, we found that it is necessary to implement two distinct
versions of the curl operator and that there is a fundamental difference between
volumetric "point" fields and "surface" fields.

(Ref. [Operators.py](https://github.com/l4m4re/fdtd/blob/master/fdtd/operators.py ))

In other words, the discrete finite difference implementation of the vector
Laplacian learns that the various fields in the model are located at specific
geometric points in space relative to one another, a detail that is lost in
infinitesimal calculus.


### Rankine ring vortex model

In order to describe an elemental ring vortex in our superfluid medium, we can use the Rankine vortex model, which is a well known model for describing vortex rings in superfluids. The Rankine vortex model assumes a solid-body rotation inside a (curved) cylinder of radius $a$ and a potential vortex outside the cylinder. The radius $a$ is referred to as the vortex-core radius. The velocity components $(v_r, v_{\theta}, v_z)$ of the Rankine vortex, expressed in terms of the cylindrical-coordinate system $(r, \theta, z)$ are given by:

$$
\displaystyle v_r = 0, \quad v_{\theta}(r) = \frac{\Gamma}{2\pi} \begin{cases} r/a^2 & r \leq a, \\ 1/r & r > a \end{cases}, \quad v_z = 0
$$
<p style="text-align: right;">(eq. 9)</p>

where $\Gamma$ is the circulation strength of the Rankine vortex. Since solid-body rotation is characterized by an azimuthal velocity $\Omega r$, where $\Omega$ is the constant angular velocity, one can also use the parameter $\Omega = \Gamma/(2\pi a^2)$ to characterize the vortex.

The vorticity field ${\displaystyle (\omega _{r},\omega _{\theta },\omega _{z})}$ associated with the Rankine vortex is

$$
    {\displaystyle \omega _{r}=0,\quad \omega _{\theta }=0,\quad \omega _{z}={\begin{cases}2\Omega &r\leq a,\\0&r>a\end{cases}}.}
$$    
<p style="text-align: right;">(eq. 10)</p>

At all points inside the core of the Rankine vortex, the vorticity is uniform at twice the angular velocity of the core; whereas vorticity is zero at all points outside the core because the flow there is irrotational. 

(Ref. [Wikipedia](https://en.wikipedia.org/wiki/Rankine_vortex) )


For the elemental ring vortex, we can use the Rankine vortex model to describe the vortex as a solid body rotation inside a ring or toroid with big toroidal radius $R$ and small poloidal radius $r$ and a potential vortex outside the ring.

For the poloial rotation, $v_{\phi}$ is given by the Rankine vortex model as $v_{\phi} = \frac{\Gamma}{2\pi} \frac{1}{r}$, and $r$ is the small poloidal radius of the ring vortex, so we can define the angular momentum of the poloidal rotation as:

$$
\displaystyle L_{\phi} = m r v_{\phi} = m r \frac{\Gamma}{2\pi} \frac{1}{r}
$$
<p style="text-align: right;">(eq. 17)</p>

Which simplifies to:

$$
\displaystyle L_{\phi} = \frac{m \Gamma}{2\pi}
$$
<p style="text-align: right;">(eq. 18)</p>

For $\Gamma = h/m$, we can define the angular momentum of the poloidal rotation as:

$$
\displaystyle L_{\phi} = \frac{m \Gamma}{2\pi} = \frac{m h}{2\pi m} = \frac{h}{2\pi}
$$
<p style="text-align: right;">(eq. 19)</p>

And we can define the angular velocity of the poloidal rotation as:

$$
\displaystyle \omega_{\phi} = \frac{v_{\phi}}{r} = \frac{\Gamma}{2\pi r^2}
$$
<p style="text-align: right;">(eq. 20)</p>

For the angular momentum of the toroidal rotation, we consider the ring as a solid disk
with mass $m$ and radius $R$, rotating with angular velocity $\omega_\theta$, so we can define the angular momentum of the toroidal rotation as:

$$
\displaystyle L_{\theta} = m R \omega_{\theta} 
$$
<p style="text-align: right;">(eq. 21)</p>

Since the Rankine vortex model applies to the poloidal rotation, we have no information on the angular velocity of the toroidal rotation, but we can assume the poloidal angular velocity to
be equal to a whole number $n$ times the toroidal angular velocity:

$$
\displaystyle \omega_{\theta} = \frac{\omega_{\phi}}{n}
$$
<p style="text-align: right;">(eq. 22)</p>

\displaystyle L_{\theta} = \frac{m \Gamma}{2\pi n \epsilon^2}

<p style="text-align: right;">(eq. 24)</p>

$$

### Linear velocity and momentum of a Vortex Ring

The velocity $\mathbf{v}$ at a point $\mathbf{r}$ due to a vortex filament with circulation $\Gamma$ is given by the Biot-Savart law:

$$
\mathbf{v}(\mathbf{r}) = \frac{\Gamma}{4\pi} \int_{\mathcal{C}} \frac{\mathbf{d\ell} \times (\mathbf{r} - \mathbf{r'})}{|\mathbf{r} - \mathbf{r'}|^3} 
$$
<p style="text-align: right;">(eq. 27)</p>

where:
- $\mathcal{C}$ is the contour of the vortex filament.
- $\mathbf{r'}$ is the position vector along the vortex filament.
- $\mathbf{d\ell}$ is the differential element along the vortex filament.

For a circular vortex ring of radius $R$, the induced velocity at a point $\mathbf{r}$ due to each differential element $\mathbf{d\ell}$ on the ring needs to be integrated over the entire ring. 

When we consider a vortex ring centered at the origin in the $xy$-plane, the velocity $v_z$ at a point $z$ on the axis of a ring vortex of radius $R$ and circulation $\Gamma$ is given by: 

$$
 v_z = \frac{\Gamma R^2}{2 (R^2 + z^2)^{3/2}} 
$$
<p style="text-align: right;">(eq. 28)</p>

This result gives you the axial component of the velocity induced by the ring vortex. For points off the axis, the velocity field would need to be computed considering the more general form of the Biot-Savart law, which typically involves more complex integrals and may require numerical methods to evaluate.

For $z = 0$, which means at the center of the ring along the axis, this simplifies to:

$$ v_z = \frac{\Gamma R^2}{2 (R^2 + 0)^{3/2}} = \frac{\Gamma R^2}{2 R^3} = \frac{\Gamma}{2R} $$
<p style="text-align: right;">(eq. 29)</p>

So, at the center of the ring (along the axis, where $z = 0$), the velocity $v_z$ is:

$$ v_z = \frac{\Gamma}{2R} $$
<p style="text-align: right;">(eq. 30)</p>

This is a straightforward result showing that the velocity induced by the vortex ring at its geometric center is directly proportional to the circulation $\Gamma$ and inversely proportional to the radius $R$ of the ring.




The volume of a torus with major radius $R$ and minor radius $r$ is given by:

$$ 
V = 2\pi^2 R r^2 
$$
<p style="text-align: right;">(eq. 31)</p>



The linear momentum $P$ is the product of the mass of the ring vortex and its velocity. The mass $m$ of the ring vortex can be calculated as:

$$ 
m = \rho \cdot V = \rho \cdot 2\pi^2 R r^2 
$$
<p style="text-align: right;">(eq. 32)</p>


Therefore, the linear momentum $P$ is:

$$ 
P = m \cdot v_z 
$$
<p style="text-align: right;">(eq. 33)</p>
   
Substituting the expressions for $m$ and $v_z$:
   
$$ 
P = (\rho \cdot 2\pi^2 R r^2) \cdot \left(\frac{\Gamma}{2R}\right) 
$$
<p style="text-align: right;">(eq. 34)</p>

Thus, the linear momentum $P$ of the ring vortex is given by:

$$ P = \rho \cdot \pi^2 r^2 \Gamma $$
<p style="text-align: right;">(eq. 35)</p>

This gives the total linear momentum of the ring vortex in terms of the fluid density $\rho$, the minor radius $r$, and the circulation $\Gamma$.

And by substituting $\Gamma = k = \frac{\eta}{\rho}$, we obtain:

$$ 
P = \pi^2 r^2 \eta 
$$
<p style="text-align: right;">(eq. 36)</p>

This result provides the linear momentum of the ring vortex in terms of the dynamic viscosity $\eta$ and the minor radius $r$ of the vortex ring. This demonstrates how the unique properties of superfluids, where the kinematic viscosity can be directly related to circulation, influence the dynamics of such vortices.

By substituting $ r = \epsilon R $, we obtain:

$$ 
P = \pi^2 \epsilon^2 R^2 \eta 
$$

\displaystyle P_l = R q.
$$
<p style="text-align: right;">(eq. 88)</p>

An interesting detail is that when this exercise is computed for a sphere with
radius $R$, we obtain the same result. This may have implications with respect to
the symmetry of the electric field, which also describes a linear component.


The area of a ring vortex with big toroidal radius $R$ and small polloidal radius $r$ is given by:

$$
\displaystyle A = 4\pi^2Rr.
$$
<p style="text-align: right;">(eq. 89)</p>

and thus the ratio R/A becomes:

$$
\displaystyle R/A = R/4\pi^2Rr = 1/4\pi^2r.
$$
<p style="text-align: right;">(eq. 90)</p>

while ratio r/A becomes:

$$
\displaystyle r/A = r/4\pi^2Rr = 1/4\pi^2R.
$$
<p style="text-align: right;">(eq. 91)</p>


With this, we can define angular momentum $P_a={\frac{m \omega}{2 \pi}}$ for a ring vortex with mass m and charge q by:

$$
\displaystyle P_a = m \text{ curl curl }\text (q/\rho) 
                  = \frac{m}{(4 \pi^2 r)(4\pi^2R)} (q/\rho).
$$
<p style="text-align: right;">(eq. 92)</p>

By substituting $m = \rho 2\pi^2 R r^2$ and simplifying, we then obtain

$$

<p style="text-align: right;">(eq. 93)</p>



The momentum of a quantized ring vortex in a superfluid is given by:

$$

$$
<p style="text-align: right;">(eq. 94)</p>

with ε the ratio of the big toroidal radius $R$ to the small polloidal radius $r$, or $ε=r/R$.

We can re-write this into two terms:

$$
\displaystyle P = \rho \Gamma \pi R^2  + 3/4 \rho \Gamma \pi ε^2 R^2
$$
<p style="text-align: right;">(eq. 95)</p>

And substite $ε=r/R$:

$$
\displaystyle P = \rho \Gamma \pi R^2  + 3/4 \rho \Gamma \pi r^2,
$$
<p style="text-align: right;">(eq. 96)</p>

to come to the assumption that the left term represents linear momentum and the right term represents angular momentum and thus we can write:

$$
\displaystyle P_l = \rho \Gamma \pi R^2
$$
<p style="text-align: right;">(eq. 97)</p>

$$
\displaystyle P_a = 3/4 \rho \Gamma \pi r^2
$$
<p style="text-align: right;">(eq. 98)</p>

This appears to offer new insight in the mechanism by which intertia works,
since both terms must be equal in magnitude, so a change in velocity or linear
momentum must be accompanyied by a change in angular momentum, and vice versa.
In other words, it appears that inertia can be attributed to the self propelling
mechanism of the ring vortex.

We can equate the linear momentum equations:

$$
\displaystyle P_l = \rho \Gamma \pi R^2 = R q
$$

<p style="text-align: right;">(eq. 99)</p>

And solve for R:

$$
\displaystyle  R = \frac{q}{\rho \Gamma \pi }

<p style="text-align: right;">(eq. 100)</p>

We can then equate the angular momentum equations:

<p style="text-align: right;">(eq. 103)</p>

And solve for r:

$$
\displaystyle r = \frac{q} { 6 \pi^3 \rho \Gamma } 
$$
<p style="text-align: right;">(eq. 104)</p>

And thus we can define angular charge $q_a$ by substituting $\Gamma = k$:

$$
\displaystyle q_a = 6 \rho k \pi^3 r
$$
<p style="text-align: right;">(eq. 105)</p>

And by substituting $\eta = \rho k$ we obtain:

$$
\displaystyle q_a = 6 \eta \pi^3 r
$$
<p style="text-align: right;">(eq. 106)</p>


This way, we can compute R and r for the elemental ring vortex with charge e and circulation $\Gamma=k$ and compute the mass of it's core, which is actually a mass displacement since the core is hollow:

### Definition of elemental charge from the fine structure constant definition

The fine structure constant α is defined by:

$$
\displaystyle α = {e^2 \over 4 π \epsilon_0 {h\over 2π} c}
$$
<p style="text-align: right;">(eq. 111)</p>

(Ref. [Wikipedia](https://en.wikipedia.org/wiki/Fine-structure_constant) )


Within our model, we substitute $\epsilon_0 = \rho$. And then we can simplify to:

$$
\displaystyle α = {e^2 \over 2 ρ h c}.
$$
<p style="text-align: right;">(eq. 112)</p>

And then we can rewrite this to define elementary charge:


$$
\displaystyle e^2 = e*e = {2 α ρ h c}.
$$
<p style="text-align: right;">(eq. 113)</p>

Next, we use this approximation of α:

$$
\displaystyle α = {1 \over 8π^2√3}.
$$
<p style="text-align: right;">(eq. 114)</p>

(Ref: eq. 16 in: [The Atomic Vortex Hypothesis, a Forgotten Path to Unification](https://vixra.org/pdf/1310.0237v1.pdf) )

However, we consider the ratio in propagation speed between longitudinal waves and transverse waves to be equal to $\pi/2$, so we substitute $\sqrt{3}$ by $π/2$ and simplify to obtain: 

$$
\displaystyle α = {1 \over 4π^3}.
$$
<p style="text-align: right;">(eq. 115)</p>



And we substite α by this approximation and simplify to obtain:

$$
\displaystyle e*e = {ρ h c \over 2π^3}.
$$
<p style="text-align: right;">(eq. 116)</p>

Next step is to multiply by $m/m$, $R/R$ and $π/π$ and re-arrange:


$$
\displaystyle e*e = ρ {h \over m} π R * { m c \over 2 π^4 R}.
$$
<p style="text-align: right;">(eq. 117)</p>

And then we evaluate both terms:

```python
angular = (m*c)/(2*pi**4*R)
linear  = rho*(h/m)*pi*R

print("linear or dielectric:", linear)
print("angular or magnetic: ", angular)
print("elemental charge    :", e)
```

This is very remarkable, since we obtain two terms that are both very close to the value for elementary charge $e$, whereby the linear term is already exactly equal to $e$. And its unit of measurement evaluates to $[kg/m^3] * [m^2/s] * [m] = [kg/s]$, the unit we assigned to electrical charge. And as can be expected, the unit of measurement for the angular term also works out to $[kg/s]$, so it seems that charge actually has two components, a rather unexpected result that offers deep insight in the actual nature of charge.

### Conclusions

By evaluting momentum for a charged particle along a descretized version of
vector Laplacian for a ring vortex with mass $m$ and charge $q$, we have shown
that electric charge actually has two components, the angular component $q_a$
and the linear component $q_l$. When we re-arranged the definition for elemental
charge squared, we found the linear term to be equal to our linear charge $q_l$,
while the angular term containing the fine structure constant $\alpha$ is not
equal to our angular charge $q_a$ even though it's value is also (almost) equal
to the value for elementary charge $e$, confirming that charge actually has two
components, a rather unexpected result.

Unfortunately, it is not easy to relate the angular term to the angular charge
we found, $q_a$, so we are unable to fully explain the physical origin of the
fine stucture constant as some eigenvalue of the ring vortex topology c.q.
resulting from the vortex ring geometry, so open questions remain around the
angular component.

A possible reason for the discrepancy is that we substituted $m = \rho 2\pi^2 R r^2$ in the momentum equations, which appears to work out fine for the linear
term, which involves (integrals over) a volume, but since the angular term
involves (integrals over) a closed curve, it is likely that the angular
derivation is not entirely correct and we should not substitue $m$ in the angular
term.
