The **acoustic wave equation** describes the propagation of small disturbances (sound waves) in a compressible medium. It is derived from the fundamental equations of fluid dynamics: the **continuity equation**, **momentum equation (Euler's equation for inviscid flow)**, and an **equation of state** (often assuming adiabatic conditions).  

---

### **1. General Form of the Acoustic Wave Equation**
For a small perturbation in pressure $ p' $ propagating through a homogeneous medium with uniform mass density $ \rho_0 $ and sound speed $ c $, the acoustic wave equation is:

$$
\frac{\partial^2 p'}{\partial t^2} = c^2 \nabla^2 p'.
$$

where:
- $ p' $ is the **acoustic pressure perturbation**,
- $ c $ is the **speed of sound**, given by $ c = \sqrt{\frac{B}{\rho_0}} $, where $ B $ is the **bulk modulus** of the medium,
- $ \nabla^2 $ is the **Laplacian operator**.

This equation describes a classic **second-order wave equation**, meaning sound waves propagate at speed $ c $ in the medium.

---

### **2. Derivation from Fluid Dynamics**
The acoustic wave equation follows from the governing equations of a compressible fluid:

1. **Continuity Equation (Mass Conservation)**  
   $$
   \frac{\partial \rho}{\partial t} + \nabla \cdot (\rho \mathbf{v}) = 0.
   $$
   For small perturbations, we assume $ \rho = \rho_0 + \rho' $, where $ \rho_0 $ is the equilibrium density and $ \rho' $ is a small fluctuation.

2. **Momentum Equation (Euler's Equation for Inviscid Flow)**
   $$
   \frac{\partial \mathbf{v}}{\partial t} + (\mathbf{v} \cdot \nabla) \mathbf{v} = -\frac{1}{\rho} \nabla p.
   $$
   Under the **linear approximation** (small perturbations), the nonlinear term $ (\mathbf{v} \cdot \nabla) \mathbf{v} $ is neglected, so it simplifies to:
   $$
   \frac{\partial \mathbf{v}'}{\partial t} = -\frac{1}{\rho_0} \nabla p'.
   $$

3. **Equation of State (Compressibility)**
   Assuming an **adiabatic process**, pressure and density perturbations are related by:
   $$
   p' = c^2 \rho'.
   $$

4. **Deriving the Wave Equation**
   - Taking the time derivative of the continuity equation:
     $$
     \frac{\partial}{\partial t} \left( \frac{\partial \rho'}{\partial t} + \rho_0 \nabla \cdot \mathbf{v}' \right) = 0.
     $$
   - Substituting $ \frac{\partial \mathbf{v}'}{\partial t} = -\frac{1}{\rho_0} \nabla p' $:
     $$
     \frac{\partial^2 \rho'}{\partial t^2} = -\rho_0 \nabla \cdot \left( -\frac{1}{\rho_0} \nabla p' \right).
     $$
     which simplifies to:
     $$
     \frac{\partial^2 \rho'}{\partial t^2} = \nabla^2 p'.
     $$
   - Using $ p' = c^2 \rho' $, we obtain the **acoustic wave equation**:
     $$
     \frac{\partial^2 p'}{\partial t^2} = c^2 \nabla^2 p'.
     $$
