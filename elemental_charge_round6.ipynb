the hollow core radius $r$ of such an elemental vortex ring particle can be
computed, as well as their ratio $\epsilon = r/R$.

### Definition of the medium

We assume all space to be filled with a superfluid like medium, which behavior can be defined by three constants: the speed of licht, the viscosity of the medium and Planck's constant. 

Hereby, we define the value of viscosity $\eta$ by the inverse of vacuum permeability  $\mu_0$, which used to be defined as:  $4\pi \times 10^{-7}$ H/m.

(Ref: [Wikipedia](https://en.wikipedia.org/wiki/Vacuum_permeability) )


----

```python
from math import *

c     = 299792458.0     # speed of light     [m/s]
eta   = 1/(4*pi*1e-7)   # viscosity          [kg/m-s],   [Pa-s] 
h     = 6.62607015e-34  # Planck's constant  [kg-m^2/s], [J-s]  
```

From here, we can define further constants, starting with mass density $\rho$, which we initially define to have a value equal to permittivity ${\epsilon_0}$, but a unit of measurement in $[kg/m^3]$ by using the definition of permittivity ${\epsilon_0}$:

$$
\displaystyle {\epsilon_0 = \frac{1}{c^2 \mu_0}}.
$$
<p style="text-align: right;">(eq. 3)</p>

And substituting $\rho = \epsilon_0$ and $\eta = 1/\mu_0$:

$$
\displaystyle {\rho = \frac{\eta}{c^2}}. 
$$
<p style="text-align: right;">(eq. 4)</p>

This comes within 10 significant digits of the SI value for permittivity ${\epsilon_0}$, also known as the electric constant. The reason for this is that the susceptibility ${\mu_0}$ has been redefined in 2019, while we prefer the classical exact definition.

From here, we can work out the unit of measurement for charge in Coulomb by equating the units of measurement of $\rho$ in $[\frac{kg}{m^3}]$ and $\epsilon_0$  in $[\frac{C^2}{N-m^2}]$ or $[\frac{C^2 \cdot s^2}{kg \cdot m^3}]$:

$$
\frac{C^2 \cdot s^2}{kg \cdot m^3} = \frac{kg}{m^3} \Rightarrow C^2 \cdot s^2 = kg^2 \Rightarrow C = \frac{kg}{s}
$$
<p style="text-align: right;">(eq. 5)</p>

Thus, charge in Coulombs can be viewed as having the unit $[\frac{kg}{s}]$, providing a new physical dimension to the concept of charge.


Then, we can define the diffusivity, kinematic viscosity or quantum circulation constant $k$, as follows:

$$
\displaystyle k = \frac{\eta}{\rho}. 
$$
<p style="text-align: right;">(eq. 6)</p>

This way, we obtain a unit of measurement in $[m^2/s]$ and a value equal to light speed c squared for $k$. However, because k is considered a fundamental constant, we change the order of these definitions a bit and define $k$ as having a value equal to $c^2$ and a unit of measurment in $[m^2/s]$ and use that to define mass density $\rho$ as follows:

$$
\displaystyle \rho  = \frac{\eta}{k}. 
$$

<p style="text-align: right;">(eq. 7)</p>

### Definition and Units of $\rho_{q0}$

The proposed definition of $\rho_{q0}$ is:

$$

- $q$ is the charge (with units $[\text{kg}/\text{s}]$),
- $m$ is mass (with units $[\text{kg}]$),

### Units of $\rho_{q0}$

Given the definitions:
1. $\frac{q}{m}$ has units of $[\text{s}^{-1}]$,
2. $\rho$ has units of $[\text{kg}/\text{m}^3]$,
3. $\eta = \rho k$ has units of $[\text{kg}/(\text{m} \cdot \text{s})]$,
4. $q$ has units of $[\text{kg}/\text{s}]$,
5. $k$ has units of $[\text{m}^2/\text{s}]$,
6. $h$ has units of $[\text{kg} \cdot \text{m}^2/\text{s}]$.

Let's check the units for each term in $\rho_{q0}$:

1. **Term 1: $\frac{q}{m} \rho$**

   $$
   \frac{q}{m} \rho = \left(\frac{\text{kg}/\text{s}}{\text{kg}}\right) \times \left(\frac{\text{kg}}{\text{m}^3}\right) = \frac{1}{\text{s}} \times \frac{\text{kg}}{\text{m}^3} = \frac{\text{kg}}{\text{m}^3 \cdot \text{s}}
   $$

2. **Term 2: $\frac{q}{h} \eta$**

   $$
   \frac{q}{h} \eta = \left(\frac{\text{kg}/\text{s}}{\text{kg} \cdot \text{m}^2/\text{s}}\right) \times \left(\frac{\text{kg}}{\text{m} \cdot \text{s}}\right) = \left(\frac{1}{\text{m}^2}\right) \times \left(\frac{\text{kg}^2}{\text{m} \cdot \text{s}^2}\right) = \frac{\text{kg}^2}{\text{m}^3 \cdot \text{s}^2}
   $$

3. **Term 3: $q \rho \frac{k}{h}$**

   $$
   q \rho \frac{k}{h} = \left(\frac{\text{kg}}{\text{s}}\right) \times \left(\frac{\text{kg}}{\text{m}^3}\right) \times \left(\frac{\text{m}^2/\text{s}}{\text{kg} \cdot \text{m}^2/\text{s}}\right) = \left(\frac{\text{kg}^2}{\text{m}^3 \cdot \text{s}}\right) \times \left(\frac{1}{\text{kg}}\right) = \frac{\text{kg}}{\text{m}^3 \cdot \text{s}}
   $$

4. **Term 4: $\frac{q \eta}{m k}$**

   $$
   \frac{q \eta}{m k} = \left(\frac{\text{kg}/\text{s}}{\text{kg}}\right) \times \left(\frac{\text{kg}}{\text{m} \cdot \text{s}}\right) \times \left(\frac{1}{\text{m}^2/\text{s}}\right) = \left(\frac{1}{\text{s}}\right) \times \left(\frac{\text{kg}}{\text{m} \cdot \text{s}}\right) \times \left(\frac{\text{s}}{\text{m}^2}\right) = \frac{\text{kg}}{\text{m}^3 \cdot \text{s}}
   $$

### Summary of Units

The vector Laplace operator describes the second spatial derivative in three dimensions, which would be the three dimensional generalization of the one dimensional $d^2/dx^2$, and has a unit of measurement in per meter squared [$/m^2$]. 

What the Laplace operator does is establish a Helmholtz decomposition of the vector field $\vec F$, whereby the term grad div $\vec F$ results in a **linear** component, the curl free or irrotational field usually denoted $\vec E$, and the term curl curl $\vec F$ results in an **angular** component, the divergence free or incompressible field usually denoted $\vec B$. And these two components may be evaluated independently of one another. 

(Ref. [Wikipedia](https://en.wikipedia.org/wiki/Helmholtz_decomposition) )

It is because of this resulting Helmholtz decomposition that the vector Laplacian actually defines a fundamental symmetry between these linear and angular components and it is that fundamental symmetry that is broken in theories that do not derive their fields directly using the vector Laplacian, which includes Maxwell's equations as well as Navier-Stokes.

The div operator in the vector Laplacian is defined as a surface integral over a
spherical volume $(4/3) \pi R^3$, divided by radius R, whereby the limit of R is
taken to zero, while the grad operator returns the change in a scalar field $f$
over a displacement dR, divided by the distance dR.

line integral over a closed curve, divided by the area enclosed by that curve,
whereby the limit of the area is taken to zero.

surface of a volume, and therefore works on a **volume** rather than a surface, while
the curl operator involves a line integral over a closed curve, and therefore
works on a **surface** rather than a volume.

### Extending to a Dual Torus Vortex

For a dual torus vortex, we need to consider the combined effect of the two rings. Given that the rings are counter-rotating and may have a differential in their angular frequencies, we can define the effective circulation as the difference in circulations of the two rings.

Let's denote:
- $ \Gamma_1 $ as the circulation of the first ring.
- $ \Gamma_2 $ as the circulation of the second ring.
- $ \Delta \Gamma = \Gamma_1 - \Gamma_2 $ as the effective circulation for the dual vortex system.

Assuming that the radii $ R $ of both rings are the same and that the circulations are close but with a small differential, the effective circulation $ \Delta \Gamma $ can be approximated by:

$$ \Delta \Gamma = \Gamma - \Gamma + \Delta \omega_t R^2 = 2 \pi R^2 \Delta \omega_t $$

Now, using the formula for the self-propelling velocity of a vortex ring, we can express the self-propelling velocity $ V $ of the dual torus vortex as:

$$ V = \frac{\Delta \Gamma}{4 \pi R} \ln \left( \frac{8R}{r_0} \right) $$

Substituting $ \Delta \Gamma = 2 \pi R^2 \Delta \omega_t $:

$$ V = \frac{2 \pi R^2 \Delta \omega_t}{4 \pi R} \ln \left( \frac{8R}{r_0} \right) $$

Simplifying:

$$ V = \frac{R \Delta \omega_t}{2} \ln \left( \frac{8R}{r_0} \right) $$

### Summary

The self-propelling velocity $ V $ of the dual torus vortex is given by:

$$ V = \frac{R \Delta \omega_t}{2} \ln \left( \frac{8R}{r_0} \right) $$

where:
- $ R $ is the radius of the vortex rings.
- $ \Delta \omega_t $ is the differential angular frequency between the two toroidal rotations.
- $ r_0 $ is the core radius of the vortex rings.

This result indicates that the self-propelling velocity of the dual torus vortex depends on the radius $ R $ of the rings, the differential angular frequency $ \Delta \omega_t $, and the logarithmic factor involving the ratio of the radius to the core size.

### Conclusions

By evaluting momentum for a charged particle along a descretized version of
vector Laplacian for a ring vortex with mass $m$ and charge $q$, we have shown
that electric charge actually has two components, the angular component $q_a$
and the linear component $q_l$. When we re-arranged the definition for elemental
charge squared, we found the linear term to be equal to our linear charge $q_l$,
while the angular term containing the fine structure constant $\alpha$ is not
equal to our angular charge $q_a$ even though it's value is also (almost) equal
to the value for elementary charge $e$, confirming that charge actually has two
components, a rather unexpected result.

Unfortunately, it is not easy to relate the angular term to the angular charge
we found, $q_a$, so we are unable to fully explain the physical origin of the
fine stucture constant as some eigenvalue of the ring vortex topology c.q.
resulting from the vortex ring geometry, so open questions remain around the
angular component.

A possible reason for the discrepancy is that we substituted $m = \rho 2\pi^2 R r^2$ in the momentum equations, which appears to work out fine for the linear
term, which involves (integrals over) a volume, but since the angular term
involves (integrals over) a closed curve, it is likely that the angular
derivation is not entirely correct and we should not substitue $m$ in the angular
term.
